```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error

plt.rc('text')
plt.rc('font', family='sans-serif')
plt.rcParams["figure.figsize"] = [30, 5]

root_dir = '/home/gridsan/sakshay/molMCTS/'

df_dft = pd.read_csv('plotting_data/generated_molecules/y6_final_dft.csv')
df_preds = pd.read_csv('plotting_data/output_dir/y6_final_dft_chemprop_weights_y6_AL_iter_1_ensemble_preds.csv')

df_preds_patent = pd.read_csv('plotting_data/output_dir/y6_final_dft_chemprop_weights_patent_ensemble_preds.csv')

dft_props = list(df_dft['gap_dft'])
pred_props = list(df_preds['gap'])

dft_props_patent = list(df_dft['gap_dft'])
pred_props_patent = list(df_preds_patent['gap'])

# Calculate RMSE for the first pair
rmse1 = np.sqrt(mean_squared_error(dft_props, pred_props))
print(f"RMSE between df_dft and df_preds: {rmse1}")

# Calculate RMSE for the second pair
rmse2 = np.sqrt(mean_squared_error(dft_props_patent, pred_props_patent))
print(f"RMSE between df_dft_patent and df_preds_patent: {rmse2}")

df_dft2 = pd.read_csv('plotting_data/generated_molecules/patent_final_dft.csv')
df_preds_random2 = pd.read_csv('plotting_data/output_dir/patent_final_dft_chemprop_weights_patent_AL_iter_1_ensemble_preds.csv')
df_preds_EI2 = pd.read_csv('plotting_data/output_dir/patent_final_dft_chemprop_weights_patent_AL_iter_2_ensemble_preds.csv')
df_preds_patent2 = pd.read_csv('plotting_data/output_dir/patent_final_dft_chemprop_weights_patent_ensemble_preds.csv')

dft_props_random2 = list(df_dft2['gap_dft'])
pred_props_random2 = list(df_preds_random2['gap'])

dft_props_EI2 = list(df_dft2['gap_dft'])
pred_props_EI2 = list(df_preds_EI2['gap'])

dft_props_patent2 = list(df_dft2['gap_dft'])
pred_props_patent2 = list(df_preds_patent2['gap'])

# Calculate RMSE for the third pair
rmse3 = np.sqrt(mean_squared_error(dft_props_patent2, pred_props_patent2))
print(f"RMSE between df_dft_patent and df_preds_patent: {rmse3}")

# Calculate RMSE for the fourth pair
rmse4 = np.sqrt(mean_squared_error(dft_props_random2, pred_props_random2))
print(f"RMSE between df_dft_random and df_preds_random: {rmse4}")

# Calculate RMSE for the fifth pair
rmse5 = np.sqrt(mean_squared_error(dft_props_EI2, pred_props_EI2))
print(f"RMSE between df_dft_EI and df_preds_EI: {rmse5}")

def plot_hexbin(pred, targ, ax, fig, dataset, bins='log', scale='log', plot_helper_lines=False, tickwidth=3, maj_tick_size=5, min_tick_size=5, rmse=None): 
    # lim_min_y = 0.5
    # lim_max_y = 2.5
    lim_min_y = 1 if dataset in ['df_dft', 'df_dft_patent'] else 0.5
    lim_max_y = 3 if dataset in ['df_dft', 'df_dft_patent'] else 3.0
    lim_min_x = 1 if dataset in ['df_dft', 'df_dft_patent'] else 0.5
    lim_max_x = 3 if dataset in ['df_dft', 'df_dft_patent'] else 3.0
    color = 'slateblue' if dataset in ['df_dft', 'df_dft_patent'] else 'darkred'
    
    ax.scatter(pred, targ, s=100, color=color)
    ax.plot((lim_min_x, lim_max_x),
            (lim_min_y, lim_max_y),
            color='#000000',
            zorder=-1,
            linewidth=1)

    ax.spines['bottom'].set_linewidth(1)
    ax.spines['top'].set_linewidth(1)
    ax.spines['left'].set_linewidth(1)
    ax.spines['right'].set_linewidth(1)

    ax.set_xlim(lim_min_x, lim_max_x)
    ax.set_ylim(lim_min_y, lim_max_y)
    ax.set_xticks(np.arange(lim_min_x, lim_max_x + 0.25, 0.5))
    ax.set_yticks(np.arange(lim_min_y, lim_max_y + 0.25, 0.5))

    ax.tick_params(axis='y', length=6, width=3, labelsize=20, pad=10, direction='in')
    ax.tick_params(axis='x', length=6, width=3, labelsize=20, pad=10, direction='in')

    ax.set_xlabel('Predicted $E_{gap}$ / eV', fontsize=20)
    
    if rmse is not None:
        ax.text(0.03, 0.9, 'RMSE: %.3f %s' % (rmse, 'eV'), transform=ax.transAxes, fontsize=20)
    if dataset in ['df_dft_patent', 'df_dft_patent2']:
        ax.set_ylabel('DFT $E_{gap}$ / eV', fontsize=20)
    
    return ax

fig, axs = plt.subplots(1, 5, sharey=False)

(ax1, ax2, ax3, ax4, ax5) = axs

plot_hexbin(np.array(pred_props_patent), np.array(dft_props_patent), ax1, fig, dataset='df_dft_patent', scale='normal', bins='normal', rmse=rmse2)
plot_hexbin(np.array(pred_props), np.array(dft_props), ax2, fig, dataset='df_dft', scale='normal', bins='normal', rmse=rmse1)
plot_hexbin(np.array(pred_props_patent2), np.array(dft_props_patent2), ax3, fig, dataset='df_dft_patent2', scale='normal', bins='normal', rmse=rmse3)
plot_hexbin(np.array(pred_props_random2), np.array(dft_props_random2), ax4, fig, dataset='df_dft_random2', scale='normal', bins='normal', rmse=rmse4)
plot_hexbin(np.array(pred_props_EI2), np.array(dft_props_EI2), ax5, fig, dataset='df_dft_EI2', scale='normal', bins='normal', rmse=rmse5)

# Adjusting y-axis range for first two plots
ax1.set_ylim(1, 3)
ax2.set_ylim(1, 3)

# Adjusting y-axis range for last three plots
ax3.set_ylim(0.5, 2)
ax4.set_ylim(0.5, 2)
ax5.set_ylim(0.5, 2)


plt.tight_layout()
plt.savefig('plotting_data/output_dir/y6_scatter_plots.svg', dpi=400)
plt.show()
```

```python
import os
import pickle
import chemprop
import umap
import pandas as pd
import numpy as np
import json
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors as rdmd
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
import matplotlib
import matplotlib.pyplot as plt
import os, sys
sys.path.insert(0, "/home/gridsan/sakshay/experiments/chemprop")
from chemprop.train.molecule_fingerprint import molecule_fingerprint

plt.rc('text')
plt.rc('font', family='sans-serif')
plt.rcParams["figure.figsize"]=[7.5,5]

# root_dir = '/home/gridsan/sakshay/molMCTS/'

arguments_fingerprints = [
    "--test_path",
    "/dev/null",
    "--preds_path",
    '/dev/null',
    "--checkpoint_dir",
    "plotting_data/checkpoints/chemprop_weights_y6_AL_iter_1",
    '--fingerprint_type', 'last_FFN'
]

args = chemprop.args.FingerprintArgs().parse_args(arguments_fingerprints)

random_sample_smiles = [[smi] for smi in pd.read_csv('plotting_data/generated_molecules/random_sample_y6_chemprop.csv')['smiles']]
expt_smiles = [[smi] for smi in pd.read_csv('plotting_data/patent_and_expt_data/OPD_acceptors_experimental_smiles_methyl_DFT.csv')['smiles_methyl']]
gen_smiles = [[smi] for smi in pd.read_csv('plotting_data/generated_molecules/y6_final_dft_in_order.csv')['best_smiles']]

def get_fingerprints(smiles):
    preds = molecule_fingerprint(args=args, smiles=smiles)
    return preds
embeddings_dict = {
    'random': get_fingerprints(random_sample_smiles),
    'expt': get_fingerprints(expt_smiles),
    'gen': get_fingerprints(gen_smiles)
}

with open('plotting_data/output_dir/embeddings_dict.pickle', 'wb') as handle:
    pickle.dump(embeddings_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
with open('plotting_data/output_dir/embeddings_dict.pickle', 'rb') as handle:
    embeddings_dict = pickle.load(handle)

new_random = np.reshape(embeddings_dict['random'], (embeddings_dict['random'].shape[0], embeddings_dict['random'].shape[1] * embeddings_dict['random'].shape[2]))
new_expt = np.reshape(embeddings_dict['expt'], (embeddings_dict['expt'].shape[0], embeddings_dict['expt'].shape[1] * embeddings_dict['expt'].shape[2]))
new_gen = np.reshape(embeddings_dict['gen'], (embeddings_dict['gen'].shape[0], embeddings_dict['gen'].shape[1] * embeddings_dict['gen'].shape[2]))

embeddings_dict = {
    'random': new_random,
    'expt': new_expt,
    'gen': new_gen
}

reducer = umap.UMAP()
gaps_dict = {'random': np.array([i[1] for i in predict()]), 'gen': np.array(predict(fname='y6_final_100_in_order.csv'))}
with open('plotting_data/output_dir/gaps_dict.pickle', 'wb') as handle:
    pickle.dump(gaps_dict, handle)

with open('plotting_data/output_dir/gaps_dict.pickle', 'rb') as handle:
    gaps_dict = pickle.load(handle)

random_samples_properties = gaps_dict['random']

fps_full = np.concatenate((new_random, new_expt, new_gen), axis=0)

print("Running PCA")
pca = PCA(n_components=2)
scaler = StandardScaler()
fps_full_scaled = scaler.fit_transform(fps_full)
Xt_full = pca.fit_transform(fps_full_scaled)
print("Finished PCA!")

with open('plotting_data/output_dir/Xt_full_pca_chemprop_latent.pickle', 'wb') as handle:
    pickle.dump(Xt_full, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open('plotting_data/output_dir/Xt_full_pca_chemprop_latent.pickle', 'rb') as handle:
    Xt_full = pickle.load(handle)

ind1 = len(new_random)
ind2 = ind1 + len(new_expt)
ind3 = ind2 + len(new_gen)
inds = [ind1, ind2, ind3]

Xt_random = Xt_full[0:inds[0]]
Xt_expt = Xt_full[inds[0]:inds[1]]
Xt_gen = Xt_full[inds[1]:inds[2]]

print("Plotting results")
fig, ax = plt.subplots(1, 1)

random_plt = ax.scatter(Xt_random[:,0], Xt_random[:,1], label='MCTS random Y6', c=random_samples_properties, cmap='viridis')
cbar = plt.colorbar(random_plt, ax=ax, pad=0.02, location='left')
for t in cbar.ax.get_yticklabels():
     t.set_fontsize(20)

cbar.ax.set_ylabel('Chemprop $E_{rep}$ / eV'.replace('rep', 'gap'), fontsize=20)

gen_properties = np.arange(Xt_gen.shape[0])  # Assuming random properties for color mapping
print(Xt_gen.shape, gen_properties)
gen_plt = ax.scatter(Xt_gen[:,0], Xt_gen[:,1], c='tomato', label='MCTS opt Y6')

expt_plt = ax.scatter(Xt_expt[:,0], Xt_expt[:,1], color="darkred", label='experiment')

ax.legend(fontsize=15, loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=1)

ax.spines['bottom'].set_linewidth(1)
ax.spines['top'].set_linewidth(1)
ax.spines['left'].set_linewidth(1)
ax.spines['right'].set_linewidth(1)

ax.tick_params(axis='y', length=0, width=0, labelsize=0, pad=0, direction='in')
ax.tick_params(axis='x', length=0, width=0, labelsize=0, pad=0, direction='in')

plt.tight_layout()
plt.savefig('plotting_data/output_dir/mcts_y6_pca_chemprop_latent.png', dpi=400)
plt.savefig('plotting_data/output_dir/mcts_y6_pca_chemprop_latent.svg', dpi=400)
plt.savefig('plotting_data/output_dir/mcts_y6_pca_chemprop_latent.pdf', dpi=400)
```
